```python
%%html
<!DOCTYPE html>
<html>
  <head>
    <title>Canvas Example</title>
    <style>
      canvas {
        border: 1px solid black;
      }
    </style>
  </head>
  <body>
    <canvas id="myCanvas" width="400" height="400"></canvas>

    <script>
      // Get the canvas element
      const canvas = document.getElementById('myCanvas');

      // Get the 2D rendering context
      const ctx = canvas.getContext('2d');

      // Set the initial position of the square
      let x = 50;
      let y = 50;

      // Set the fill color
      ctx.fillStyle = 'red';

      // Draw the square at the current position
      function drawSquare() {
        ctx.clearRect(0, 0, canvas.width, canvas.height); // Clear the canvas
        ctx.fillRect(x, y, 50, 50); // Draw the square
      }

      // Update the square's position based on arrow key press
      function moveSquare(event) {
        switch (event.keyCode) {
          case 37: // Left arrow key
            x -= 10;
            break;
          case 38: // Up arrow key
            y -= 10;
            break;
          case 39: // Right arrow key
            x += 10;
            break;
          case 40: // Down arrow key
            y += 10;
            break;
        }
        drawSquare(); // Redraw the square at the new position
      }

      // Add event listener for arrow key press
      document.addEventListener('keydown', moveSquare);
      
      // Draw the initial square
      drawSquare();
    </script>
  </body>
</html>

```
